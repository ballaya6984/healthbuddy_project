```python
# Install necessary libraries
!pip install --upgrade xgboost

import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
import joblib

# Download stopwords if not already done
nltk.download('stopwords')

# Define target variable
# Assuming `df` is your DataFrame with columns `Vitamin_A`, `Vitamin_B12`, etc.
# Ensure `df` is already loaded
y = df[['Vitamin_A', 'Vitamin_B12', 'Vitamin_C', 'Vitamin_D', 'Vitamin_E', 'Vitamin_K']]

# Preprocessing function for the food names
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))

    # Step 1: Clean the food_name column (Remove non-alphabet characters)
    cleaned_text = text.str.replace(r'[^a-zA-Z\s,]', '', regex=True)
    cleaned_text = cleaned_text.str.lower()

    # Step 2: Tokenize and remove stopwords
    tokens = cleaned_text.str.split('[,\s]+')
    tokens = tokens.apply(lambda x: [word for word in x if word not in stop_words])

    # Step 3: Expand abbreviations
    abbreviation_mapping = {
        "CKD": "Cooked",
        "STMD": "Steamed",
        "RAW": "Raw",
        "SYRUP": "Syrup",
        "CANE": "Cane"
    }

    def expand_abbreviations(row_text):
        words = row_text.split(',') if isinstance(row_text, str) else []
        return ', '.join([abbreviation_mapping.get(word.upper(), word) for word in words])

    expanded_text = tokens.apply(lambda x: ' '.join(x)).apply(expand_abbreviations)
    return expanded_text

# Preprocess the food names
df['food_name_expanded'] = preprocess_text(df['food_name_expanded'])

# Initialize CountVectorizer
count_vectorizer = CountVectorizer()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    df['food_name_expanded'], y, test_size=0.2, random_state=42
)

# Initialize the XGBoost Regressor
xg_reg = xgb.XGBRegressor(random_state=42)

# Wrap the XGBoost Regressor with MultiOutputRegressor
multi_output_xg = MultiOutputRegressor(xg_reg)

# Define the hyperparameter grid for XGBoost
param_dist = {
    'estimator__n_estimators': [100, 200, 300],
    'estimator__learning_rate': [0.01, 0.05, 0.1, 0.2],
    'estimator__max_depth': [3, 5, 7, 9],
    'estimator__subsample': [0.6, 0.8, 1.0],
    'estimator__colsample_bytree': [0.6, 0.8, 1.0]
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    multi_output_xg, param_distributions=param_dist, n_iter=50,
    scoring='neg_mean_squared_error', cv=3, verbose=1, random_state=42, n_jobs=-1
)

# Create the pipeline with preprocessing and model
pipeline = Pipeline([
    ('vectorizer', count_vectorizer),  # Preprocessing step
    ('model', random_search)  # Model and Hyperparameter Tuning
])

# Train the pipeline
pipeline.fit(X_train, y_train)

# Extract the best model and hyperparameters
best_xg_model = random_search.best_estimator_
print(f"Best Hyperparameters: {random_search.best_params_}")

# Make predictions
y_pred_best_xg = best_xg_model.predict(count_vectorizer.transform(X_test))
mse_best_xg = mean_squared_error(y_test, y_pred_best_xg)
print(f"Tuned XGBoost Regressor - MSE: {mse_best_xg:.4f}")

# Save the entire pipeline (preprocessing + model) into a single .pkl file
joblib.dump(pipeline, 'preprocessing_and_tuned_xgboost_model.pkl')

print("Preprocessing and XGBoost model saved into 'preprocessing_and_tuned_xgboost_model.pkl'!")
```
